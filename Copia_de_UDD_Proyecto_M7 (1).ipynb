# preprocesamiento


```python
from sklearn.preprocessing import LabelEncoder
# Convertir las variables categóricas a variables dummy
df_dumi = pd.get_dummies(df.loc[:, ~df.columns.isin(['How likely is that you will work for one employer for 3 years or more ?'])])

# Convertir las etiquetas de clase a valores numéricos
le = LabelEncoder()
df_dumi['clase_encoded'] = le.fit_transform(df['How likely is that you will work for one employer for 3 years or more ?'])
```

```python
df_dumi['clase_encoded']
```

# modelo

vamos hacer un modelo de clasificacion multiple que nos diga si alguien de la generacion z trabajaria 3 años en una empresa, esto dado que esta generacion es conocida por migrar rapidamente de una empresa.

```python
# Importar las bibliotecas necesarias
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


# Dividir los datos en características (X) y etiquetas (y)
X = df_dumi.drop(['clase_encoded'], axis=1)
y = df_dumi['clase_encoded']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Definir los hiperparámetros a ajustar
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Crear el modelo de Random Forest
rf_model = RandomForestClassifier(random_state=42)

# Configurar la cuadrícula de búsqueda
grid_search = GridSearchCV(rf_model, param_grid, cv=5, scoring='accuracy')

# Realizar la búsqueda en la cuadrícula
grid_search.fit(X_train, y_train)

# Obtener el mejor modelo de la búsqueda en la cuadrícula
best_rf_model = grid_search.best_estimator_
print(f"Los mejores parametros son {grid_search.best_estimator_}")
# Realizar predicciones en el conjunto de prueba
y_pred = best_rf_model.predict(X_test)

# Evaluar el rendimiento del modelo
# Calcular y mostrar la precisión, recall y F1-Score
tipos =['micro', 'macro', 'weighted']
for i in tipos:
  precision = precision_score(y_test, y_pred, average=i)  # Puedes usar 'micro', 'macro', 'weighted' o 'None'
  recall = recall_score(y_test, y_pred, average=i)
  f1 = f1_score(y_test, y_pred, average=i)
  print(f"Precisión {i}: {precision}")
  print(f"Recall {i}: {recall}")
  print(f"F1-Score {i}: {f1}")
  print("----------------------------------------------------------------------------------")

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
# Imprimir el informe de clasificación
print(classification_report(y_test, y_pred))

# Calcular y mostrar la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confusión:")
print(conf_matrix)
```

```python
from flask import Flask, request, jsonify
# Implementar la API REST con Flask
app = Flask(__name__)

@app.route('/predict', methods=['POST'])
def predict():
    data = request.get_json()
    input_data = pd.DataFrame(data, index=[0])
    input_data = pd.get_dummies(input_data, columns=['variable_categorica1', 'variable_categorica2'])
    prediction = best_rf_model.predict(input_data)[0]
    return jsonify({'prediction': int(prediction)})

if __name__ == '__main__':
    app.run(debug=True)
```
